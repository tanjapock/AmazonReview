```python
from pathlib import Path
import pandas as pd
import numpy as np
import re
import spacy
nlp = spacy.load("en_core_web_sm", disable=["ner", "parser"])
```

```python
#path to preprocessed dataset
path = Path(r"C:\Users\marta\My drive (marta.infante.baiao@gmail.com) (1)\DTU\Computational Tools\Books_rating_cleaned_new.csv")
df = pd.read_csv(path)
```

```python
print("Number of columns:", df.shape[1]) #number of columns
print("Column names:", df.columns.tolist()) #name of columns
print(df['cleanText'].sample(1).iloc[0])
df.shape
```

## Lowercase

```python
df['cleanText'] = df['cleanText'].str.lower()
print(df['cleanText'].sample(1).iloc[0])
```

## Remove Ponctuation

```python
df['cleanText']=df['cleanText'].str.replace(r'[^\w\s]', '', regex=True)
print(df['cleanText'].sample(1).iloc[0])
```

## Remove Numbers

```python
df['cleanText'] = df['cleanText'].str.replace(r'\d+', '', regex=True)
print(df['cleanText'].sample(1).iloc[0])
```

## Remove non-ASCII characters

```python
df['cleanText'] = df['cleanText'].str.encode('ascii', 'ignore').str.decode('ascii')
print(df['cleanText'].sample(1).iloc[0])
```

## Remove extra white spaces

```python
df['cleanText'] = df['cleanText'].str.replace(r'\s+', ' ', regex=True).str.strip()
print(df['cleanText'].sample(1).iloc[0])
```

## Lemmatization

```python
texts = df["cleanText"].astype(str).tolist()

lemmatized = []
for doc in nlp.pipe(texts, batch_size=1000, n_process=4):  # parallel processing
    lemmas = [token.lemma_ for token in doc if not token.is_punct and not token.is_space]
    lemmatized.append(" ".join(lemmas))

df["lemmatized"] = lemmatized

print(df['cleanText'].sample(1))
```

## Check 
Check if there are any ponctuation, numbers, extra white spaces, non-ASCII characters or uppercase

```python
check = {
    "numbers": int(df['cleanText'].str.contains(r'\d', regex=True, na=False).sum()),
    "punctuation": int(df['cleanText'].str.contains(r'[^\w\s]', regex=True, na=False).sum()),
    "non_ascii": df[df['cleanText'].apply(lambda x: isinstance(x, str) and not x.isascii())].shape[0],
    "extra_spaces": int(df['cleanText'].str.contains(r'\s{2,}', regex=True, na=False).sum()),
    "uppercase": int(df['cleanText'].str.contains(r'[A-Z]', regex=True, na=False).sum())
}

print(check)
```
